### Codigo 1 ###

Se tienen un objetos de la realidad, el punto tal y como se muestra a continuación:

![punto](https://github.com/linamariaum/intro_python_actvidad1/blob/master/punto.jpg?raw=1)

El siguiente diagrama UML modela el **punto** como una clase llamada ```Point```:

![uml_punto1](https://github.com/linamariaum/intro_python_actvidad1/blob/master/uml_punto1.jpg?raw=1)

* El constructor de la clase inicializa las coordenadas de esta (x,y) en el valor asignado por el usuario si se le pasan argumentos o en 0 si no se le pasan argumentos
* Para esta clase se definieron los siguientes métodos:
  * **distance**: Calcula la distancia de un punto a otro.
  * **set_location**: Asigna una posición determinada al punto.
  * **distance_from_origin**: Calcula la distancia del punto al origen de coordenadas.
  * **print**: Imprime las coordenadas siguiendo el siguiente formato: (x, y)
  
Dado el diagrama UML anterior hacer lo siguiente:
1. Codificar la clase en python.
2. Definir tres objetos tipo ```Point```: ```A(0,0)```,  ```B(-3,4)```, ```C(5,8)```.
3. Imprimir los tres puntos anteriormente creados.
4. Cambiar las coordenadas de A de tal modo que la componentes x de A sea la componente x de C y la componente y de A sea la componente y de B.
5. Imprimir nuevamente los puntos.
6. Calcular las distancias de AB, BC, AC y almacenarlas en las variables d_AB, d_BC y d_AC respectivamente. Luego, imprimir estos valores.
7. Calcule la distancia de A hasta el origen e imprimala.


````python
"""
Punto 1: Definicion de la clase punto
"""

from math import sqrt

# Clase Point
class Point:  
    
    # Constructor  
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    # Metodos  
    def set_location(self, x, y):
        self.x = x    
        self.y = y  
        
    def distance_from_origin(self):
        return sqrt(self.x * self.x + self.y * self.y)
    
    def distance(self, other):    
        dx = self.x - other.x    
        dy = self.y - other.y    
        return sqrt(dx**2 + dy**2)
    
    def print(self):
        print("(" + str(self.x) + ", " + str(self.y) + ")")
        

"""
Punto 2: A(0,0), B(-3,4), C(5,8)
"""
A = Point()
B = Point(-3,4)
C = Point(5,8)

"""
Punto 3: Impresion de los puntos A, B, C
"""
print("Coordenadas de los puntos A, B y C")
print("A:")
A.print()
print("B:")
B.print()
print("C:")
C.print()

"""
Punto 4: A.x = C.x y A.y = B.y
"""
A.set_location(C.x,B.y)

"""
Punto 5: Impresion de los puntos A, B, C
"""
print()
print("Nuevas coordenadas de los puntos A, B y C")
print("A:")
A.print()
print("B:")
B.print()
print("C:")
C.print()

"""
Punto 6: Calculo e impresión de las distancias
"""
d_AB = A.distance(B)
d_BC = B.distance(C)
d_AC = A.distance(C)
print()
print("Distancia entre A y B: " + str(d_AB))
print("Distancia entre B y C:",d_BC)
print("Distancia entre C y A: %s" %(d_AC))

"""
Punto 7: Calculo de la distancia de A hasta el origen
"""
print()
print("Distancia entre A y el origen es: " + str(A.distance_from_origin()))
````

````python
'''Solución Punto 1: Temperatura'''

class Temperature:

  # Metodos no implementados

  def __init__(self, temperature = 75):
    self.temperature = temperature

  def __str__(self):
    # retonar de la forma "75 degrees Farenheit"
    s = str(self.temperature) + " degrees"
    return s

  def aboveFreezing(self):
    #Retorna True si la temperatura está arriba del punto de congelación (0°C, 32°F o 273.15°K)
    pass
  
  def convertToFahren(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Fahrenheit.
    pass

  def convertToCelcius(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Celcius.
    pass

  def convertToKelvin(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Kelvin.
    pass
 

''' Clase Fahrenheit '''
class Fahrenheit(Temperature):

  def __init__(self, temperature = 75):
    super().__init__(temperature)

  def __str__(self):
    # retonar de la forma "75 degrees Farenheit"
    s = super().__str__() + " Fahrenheit"
    return s

  def aboveFreezing(self):
    #Retorna True si la temperatura está arriba del punto de congelación (0°C, 32°F o 273.15°K)
    if (self.temperature > 32):
      return True
    else:
      return False
  
  def convertToFahren(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Fahrenheit.
    return Fahrenheit(self.temperature)

  def convertToCelcius(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Celcius.
    celcius = (self.temperature - 32) * 5/9
    return Celcius(celcius)

  def convertToKelvin(self):
    #Retorna un nuevo objeto tipo temperatura convertido a grados Kelvin.
    kelvin = (self.temperature - 32) * 5/9 + 273.15
    return Kelvin(kelvin)


''' Clase Celcius '''
class Celcius(Temperature):

  def __init__(self, temperature = 75):
    super().__init__(temperature)   

  def __str__(self):
    # retonar de la forma "75 degrees Celcius"
    s = super().__str__() + " Celcius"
    return s

  def aboveFreezing(self):
    #Retorna True si la temperatura está arriba del punto de congelación (0°C, 32°F o 273.15°K)
    if (self.temperature > 0):
      return True
    else:
      return False
  
  def convertToFahren(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Fahrenheit.
    fahrenheit = (self.temperature  * 9/5) + 32
    return Fahrenheit(fahrenheit)

  def convertToCelcius(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Celcius.
    return Celcius(self.temperature)

  def convertToKelvin(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Kelvin.
    kelvin = self.temperature + 273.15
    return Kelvin(kelvin)


''' Clase Kelvin '''
class Kelvin(Temperature):

  def __init__(self, temperature = 75):
    super().__init__(temperature)

  def __str__(self):
    # retonar de la forma "75 degrees Kelvin"
    s = super().__str__() + " Kelvin"
    return s

  def aboveFreezing(self):
    #Retorna True si la temperatura está arriba del punto de congelación (0°C, 32°F o 273.15°K)
    if (self.temperature > 273.15):
      return True
    else:
      return False
  
  def convertToFahren(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Fahrenheit.
    fahrenheit = (self.temperature - 273.15) * 9/5 + 32
    return Fahrenheit(fahrenheit)

  def convertToCelcius(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Celcius.
    celcius = self.temperature - 273.15
    return Celcius(celcius)

  def convertToKelvin(self): 
    #Retorna un nuevo objeto tipo temperatura convertido a grados Kelvin.
    kelvin = self.temperature
    return Kelvin(kelvin)
````

## Diagrama UML para la clase Fraccion


| Nombre de la clase |Atributos                |Métodos                       |
|----------------|-------------------------------|-----------------------------|
|Fraccion| + numerador :`Int`   | __init__(numerador: `Int`, denominador: `Int`)|
|        |+ denominador :`Int`  | __add__(other:`Fraccion`): `Fraccion`|
|        |+ fraccion :`Fraction`| __sub__(other:`Fraccion`): `Fraccion`|
|        |                      | __mul__(other:`Fraccion`): `Fraccion`|
|        |                      | __truediv__(other:`Fraccion`): `Fraccion`|
|        |                      | __str__(void): `void`|


````python
'''Solución Punto 2: Fraccion'''
from fractions import Fraction

class Fraccion:

  def __init__(self, numerador: int, denominador: int):
    self.numerador = numerador
    self.denominador = denominador
    self.fraccion = Fraction(self.numerador, self.denominador)

  def __add__(self, other):
    # operador +
    fraccion = Fraction(self.numerador, self.denominador) + Fraction(other.numerador, other.denominador)
    return fraccion

  def __sub__(self, other):
    # operador -
    fraccion = Fraction(self.numerador, self.denominador) - Fraction(other.numerador, other.denominador)
    return fraccion

  def __mul__(self, other):
    # operador *
    fraccion = Fraction(self.numerador, self.denominador) * Fraction(other.numerador, other.denominador)
    return fraccion

  def __truediv__(self, other):
    # operador /
    fraccion = Fraction(self.numerador, self.denominador) / Fraction(other.numerador, other.denominador)
    return fraccion

  def __str__(self):
    s = str(self.fraccion)
    return s


'''Haciendo uso de la Clase Fraccion'''
n: int = 10
d: int = 18
F = Fraccion(n,d)
Q = Fraccion(n,d)
print("Fraccion: "+str(F))
print("Suma: "+str(F)+" + "+str(Q)+" = "+str(F.__add__(Q)))
print("Resta: "+str(F)+" - "+str(Q)+" = "+str(F.__sub__(Q)))
print("Multiplicación: "+str(F)+" * "+str(Q)+" = "+str(F.__mul__(Q)))
print("División: ("+str(F)+") / ("+str(Q)+") = "+str(F.__truediv__(Q)))
````
